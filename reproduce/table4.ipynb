```python
import csv
import json
from pathlib import Path

# Update these two paths to your actual mounted datasets
train_root = Path("/kaggle/input/datasets/zfturbo/audioset")
valid_root = Path("/kaggle/input/datasets/zfturbo/audioset-valid")

# Example expected structure:
# train_root / "train_wav", "train.csv", "class_labels_indices.csv"
# valid_root / "valid_wav" (or similar), "valid.csv" (or balanced_train_segments.csv etc.)

out_root = Path("/kaggle/working/audioset_manifest")
out_root.mkdir(parents=True, exist_ok=True)
out_file = out_root / "audio.jsonl"

# MID -> human label
mid_to_name = {}
labels_csv = train_root / "class_labels_indices.csv"
with labels_csv.open("r", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    for row in reader:
        mid_to_name[row["mid"]] = row["display_name"]

def build_wav_index(wav_dir: Path):
    idx = {}
    if wav_dir.exists():
        for p in wav_dir.rglob("*.wav"):
            idx[p.stem] = p
    return idx

# Adjust valid wav folder name if needed
train_wav_idx = build_wav_index(train_root / "train_wav")
valid_wav_idx = build_wav_index(valid_root / "valid_wav")

def append_from_csv(csv_path: Path, wav_idx: dict, handle):
    if not csv_path.exists():
        return
    with csv_path.open("r", encoding="utf-8") as f:
        reader = csv.reader(f)
        for row in reader:
            if not row or row[0].startswith("#"):
                continue
            # AudioSet CSV columns: YTID, start_seconds, end_seconds, positive_labels
            ytid = row[0].strip()
            if ytid not in wav_idx:
                continue
            mids = [m.strip() for m in row[3].split(",") if m.strip()] if len(row) > 3 else []
            labels = [mid_to_name.get(m, m) for m in mids]
            rec = {
                "key": ytid,
                "path": str(wav_idx[ytid]),
                "labels": labels
            }
            handle.write(json.dumps(rec, ensure_ascii=False) + "\n")

with out_file.open("w", encoding="utf-8") as f:
    append_from_csv(train_root / "train.csv", train_wav_idx, f)
    append_from_csv(valid_root / "valid.csv", valid_wav_idx, f)

print("Wrote:", out_file)
```
